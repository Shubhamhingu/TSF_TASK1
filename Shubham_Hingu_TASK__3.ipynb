**Importing all the required libraries of python**

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
%matplotlib inline
```

**Importing the Data set from sklearn.datasets library**

```
data=pd.read_csv('Iris.csv')
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
y=iris.target
print(df.head())
print(y)
```

**Setting the Id column as index to access rows to avoid confusion in actual indexes and Id column**

```
data.set_index('Id',inplace=True)
```

```
data.head()
```

**Checking the count of the different species in the dataset**

```
sns.countplot('Species',data=data);
```

From Sklearn library importing the Decision Tree Classifier

```
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
print('Decision Tree Classifer Created')
```

```
x=[[5.1,3.5,1.4,0.2]]
dtree.predict(x)
```

Importing the necessary Libraries for constructing Decision tree and Constructing the tree.

```
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names, filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
